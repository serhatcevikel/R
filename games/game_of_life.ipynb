```r
neigh_count <- function(x,y, aa) sum(aa[((-1):1)+x, ((-1):1) + y ]) - aa[x,y]
neigh_countv <- Vectorize(neigh_count)
```

```r
iterate_conway <- function(dimm = 30, waitt = 0.5, maxit = 50)
{
    mm <- matrix(sample(0:1, dimm^2, replace = T), nrow = dimm)
    mm <- rbind(F, mm, F)
    mm <- cbind(F, mm, F)
    
    for (i in 1:maxit)
    {
        mm2 <- outer(2:(dimm + 1), 2:(dimm + 1), neigh_countv, mm)
        mm[-c(1, nrow(mm)), -c(1, ncol(mm))] <- mm2 %in% 2:3
        plot(rep(1:(dimm + 2), each = (dimm + 2)),
             rep(-(1:(dimm + 2)), (dimm + 2)),
             pch = ifelse(mm, 15, 0),
             cex = 2,
             asp = 1,
             xpd = NA,
             axes = FALSE,
             ann = FALSE)
        Sys.sleep(waitt)
    }
}
```

```r
iterate_conway()
```
